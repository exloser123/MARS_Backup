```python
import numpy as np

corL_dict = np.load('database/hirise/corL_dict.npy', allow_pickle=True).item()
rmsH_dict = np.load('database/hirise/rmsH_dict.npy', allow_pickle=True).item()
```

```python
corL_dict_keys = list(corL_dict.keys())
rmsH_dict_keys = list(rmsH_dict.keys())
```

```python
sample = corL_dict[corL_dict_keys[0]]
sample = sample[(sample > 0) & (sample < 20)]
from matplotlib import pyplot as plt
from scipy.stats import gaussian_kde

if len(sample) == 0:
    print('empty')
else:
    kde = gaussian_kde(sample)
    x = np.linspace(0, 20, 500)
    plt.plot(x, kde(x))
```

```python
from matplotlib import pyplot as plt
from scipy.stats import gaussian_kde
for i in range(len(corL_dict_keys)):
    sample = rmsH_dict[rmsH_dict_keys[i]]
    if len(sample) == 0:
        print(rmsH_dict_keys[i])
    else:
        kde = gaussian_kde(sample)
        x = np.linspace(0, 3, 90)
        plt.plot(x, kde(x))
plt.title('All units rmsH gaussian kde')
plt.xlabel('rmsH/m')
plt.ylabel('density')
plt.savefig('database/hirise/img/rmsH_kde.png', dpi=300)
```

```python
from matplotlib import pyplot as plt
from scipy.stats import gaussian_kde
for i in range(len(corL_dict_keys)):
    sample = corL_dict[corL_dict_keys[i]]
    if len(sample) == 0:
        print('empty')
    else:
        kde = gaussian_kde(sample)
        x = np.linspace(0, 20, 100)
        plt.plot(x, kde(x))
plt.title('All units corL gaussian kde')
plt.xlabel('corL/m')
plt.ylabel('density')
plt.savefig('database/hirise/img/corL_kde.png', dpi=300)
```
