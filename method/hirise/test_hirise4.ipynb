```python
import numpy as np
PRODUCT_ID = 'DTEEC_002786_1580_002074_1580_A01.IMG'
PRODUCT_ID = PRODUCT_ID[:-4]
corL_path = 'data/result/'+PRODUCT_ID+'_corL.npy'
rmsH_path = 'data/result/'+PRODUCT_ID+'_rmsH.npy'
corL_data = np.load(corL_path,allow_pickle=True)
rmsH_data = np.load(rmsH_path,allow_pickle=True)
```

```python
from matplotlib import pyplot as plt
plt.hist(corL_data,bins=1000)
# y轴title
plt.ylabel('Number')
# x轴title
plt.xlabel('corL (m)')
plt.savefig(PRODUCT_ID+'_corL.png')
```

```python
plt.hist(rmsH_data,bins=1000)
# y轴title
plt.ylabel('Number')
# x轴title
plt.xlabel('rmsH (m)')
plt.savefig(PRODUCT_ID+'_rmsH.png')
```
